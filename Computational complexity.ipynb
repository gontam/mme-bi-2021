Suppose that an algorithm, running on an input of size N, takes 10N^2 + 100N + 1000 machine instructions.
The 10N^2 becomes larger than the remaining terms, 100N + 1000, once N becomes large enough.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(range(100))
y1 = 10*x**2
y2 = 100*x + 1000
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

In this case, we could assume that the running time of this algorithm grows as N^2. We drop the coefficient 100 and the remaining 100N + 1000. We assume that, in case our algorithm under aN^2 + b*N + c conditions, there will always be a value N for which a*N^2 is greater than b*N + c and with greater N this difference increases.
